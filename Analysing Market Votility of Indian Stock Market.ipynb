The code bnf['Prev. Close'] = bnf['Close'].shift() is used to create a new column in the DataFrame bnf (presumably containing financial data) called 'Prev. Close.' This new column will contain the values of the 'Close' column shifted by one position (one row up). In other words, each row in the 'Prev. Close' column will contain the 'Close' value from the previous row.

```python
bnf['Prev. Close']=bnf['Close'].shift()
```

```python
bnf['Prev. Close'].fillna(18177.20, inplace = True)
bnf.reset_index(drop=True, inplace=True)
```

After executing this code, the bnf DataFrame will have a new column called "Gap," which contains the calculated opening price gaps for each trading day. These gaps represent the difference between how Bank Nifty opened on a given day and how it closed on the previous trading day. These gap values can be used for further analysis, such as identifying gap trading strategies or assessing market sentiment at the opening of each trading session.






```python
bnf['Gap'] = bnf['Open'] - bnf['Prev. Close']
bnf['% Gap'] = round((bnf['Open']-bnf['Prev. Close'])/bnf['Prev. Close']*100,2)
bnf['Abs % Gap'] = abs(round((bnf['Open']-bnf['Prev. Close'])/bnf['Prev. Close']*100,2))
```

I gap is greater than '+-3/+-2/+-1' print true else False

```python
def three_percent(num):
    if num>=3:
        return("False")
    else:
        return('True')
```

```python
def two_percent(num):
    if num>=2:
        return("False")
    else:
        return('True')
```

```python
def one_percent(num):
    if num>=1:
        return("False")
    else:
        return('True')
```

```python
bnf['+-3% Gap']= bnf['Abs % Gap'].apply(lambda x: three_percent(x))
bnf['+-2% Gap']= bnf['Abs % Gap'].apply(lambda x: two_percent(x))
bnf['+-1% Gap']= bnf['Abs % Gap'].apply(lambda x: one_percent(x))
```
